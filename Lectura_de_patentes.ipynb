```python
direccion = 'https://http2.mlstatic.com/D_NQ_NP_945063-MLA27837044837_072018-O.webp'
direccion2 = 'https://d2r9epyceweg5n.cloudfront.net/stores/001/097/145/products/chapa-patente-antigua-1900-21-9d80dd03df62ac40db16110035470326-1024-1024.jpg'
im= Image.open(requests.get(direccion, stream=True).raw).convert('L')
im 
```

```python

import keras_ocr 
import pandas as pd
pipeline = keras_ocr.pipeline.Pipeline()
  
images = [
    keras_ocr.tools.read(url) for url in [
        
        
        direccion, direccion2

       
    ]
  ]
  
#print(images[0])
#print(images[1])

prediction_groups = pipeline.recognize(images)

predicted_image = prediction_groups[0]
pd.DataFrame(prediction_groups)
#for text, box in predicted_image:
 #   print(text)
```
