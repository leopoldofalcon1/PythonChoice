```python
from PIL import Image
import requests
url = "https://i.imgur.com/J6bfBqz.png"

im = Image.open(requests.get(url, stream=True).raw)
im
```

```python
tran = im.transpose(0)
tran = tran.effect_spread(10)
tran
```
